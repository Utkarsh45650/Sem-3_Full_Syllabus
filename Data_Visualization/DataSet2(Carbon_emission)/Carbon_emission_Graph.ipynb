```python
import pandas as pd
import matplotlib.pyplot as plt
```

Basic Graph

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_productA = [1500, 1800, 2200, 2500, 3000, 2000, 4000, 4200, 3000, 4500, 2000, 3500]
sales_productB = [3200, 3600, 3100, 2300, 3800, 4900, 2800, 2000, 1900, 2300, 3800, 1300]
plt.plot(months, sales_productA, marker='o', label='Product A', color='blue')
plt.plot(months, sales_productB, marker='o', label='Product B', color='red', linewidth=2, linestyle='-.')
plt.title('Monthly Sales Data of Product A and B')
plt.xlabel('Months', fontdict={'fontsize': 12, 'fontweight': 'bold', 'color': 'green', 'style': 'italic', 'family': 'serif'})
plt.ylabel('Sales (in USD)', fontdict={'fontsize': 12, 'fontweight': 'bold', 'color': 'green', 'style': 'italic', 'family': 'serif'})
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', linewidth=0.5, axis='y', color='gray')
plt.show()
```

Graph on DataSet2


```python
#pip install openpyxl  run this command befor loding the excel file. 
df = pd.read_excel('Carbon_emission(small DataSet2).xlsx')
df
```

```python
plt.plot(df['Year'], df['India'], label='India', color='orange', marker='o')
plt.plot(df['Year'], df['China'], label='China', color='red', marker='o')
plt.plot(df['Year'], df['USA'], label='USA', color='green', marker='o')
plt.title('Carbon Emission by Country')
plt.xlabel('Year')
plt.ylabel('Carbon Emission (in million tonnes)')
plt.legend()
plt.show()
```

Simple way to plot two graphs side by side using subplots

```python
#in this there is indexing not numbering
fig, (India, USA) = plt.subplots(1, 2, figsize=(12, 6))
India.plot(df['Year'], df['India'], label='India', color='orange', marker='o')
USA.plot(df['Year'], df['USA'], label= 'USA', color= 'red', marker= 'o')
India.set_title('Carbon Emission in India')
USA.set_title('Carbon Emission in USA')
fig.suptitle('Carbon Emission Comparison')
```

Another way plot graph side by side

```python
#in this there is numbers not indexing
plt.subplot(1,2,1)
plt.plot(df['Year'], df['India'], label='India', color='orange', marker='o')
plt.title('Carbon Emission in India')
plt.subplot(1,2,2)
plt.plot(df['Year'], df['USA'], label= 'USA', color= 'red', marker= 'o')
plt.title('Carbon Emission in USA')
plt.suptitle('Carbon Emission Comparison')
```
