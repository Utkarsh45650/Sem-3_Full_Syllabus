```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ['Electronics', 'Clothing', 'Home', 'Books', 'Toys', 'Furniture', 'Sports']
sales = [200, 150, 300, 100, 250, 400, 350]

plt.bar(categories, sales, color= 'skyblue')
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.savefig('bar_chart.png')  # Save the figure as a PNG file
plt.show()
#don't use plt.savefig() if you don't want to save the figure
#don't call plt.savefig() after plt.show() as it won't have any effect
```

Creating clustered bar graph

```python
#Data
employee_name = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace']
employee_productivity = [80, 90, 70, 85, 95, 60, 75]
employee_attendance = [22, 20, 18, 21, 23, 19, 20]
employee_satisfaction = [4.5, 4.0, 3.5, 4.2, 4.8, 3.8, 4.1]

#X-Axis setup
x = np.arange(len(employee_name))
width = 0.25  # Width of the bars

# Create bars for each category
plt.bar(x - width, employee_productivity, width=width, label='Productivity', color='lightblue')
plt.bar(x, employee_attendance, width=width, label='Attendance', color='lightgreen')
plt.bar(x + width, employee_satisfaction, width=width, label='Satisfaction', color='lightcoral')

# Add labels and title
plt.xlabel('Employee Names')
plt.ylabel('Scores')
plt.title('Employee Performance Metrics')
plt.xticks(x, employee_name)
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Creating Stack bar graph

```python
# Create bars for each category
plt.bar(x, employee_productivity, width=width, label='Productivity', color='lightblue', bottom=0)
plt.bar(x, employee_attendance, width=width, label='Attendance', color='lightgreen', bottom=employee_productivity)
plt.bar(x, employee_satisfaction, width=width, label='Satisfaction', color='lightcoral', bottom=np.array(employee_productivity) + np.array(employee_attendance))
#use np.array() for element wise addition
#don't use employee_attendance + employee_productivity as it expend the list

# Add labels and title
plt.xlabel('Employee Names')
plt.ylabel('Scores')
plt.title('Employee Performance Metrics')
plt.xticks(x, employee_name)
plt.legend()

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
